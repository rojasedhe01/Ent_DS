```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(3,4))
ax.hist(x= df.work_year, density=False, orientation='vertical')
ax.set_xlabel('Año')
ax.set_ylabel('Cantidad')
ax.set_title('Historico Por Año (Colombia)')
```

### Historico Por Año (Colombia)
#### Analizando el grafico anterior podemos identificar el crecimiento laboral por año

```python
import seaborn as sns 
serie=df['experience_level'].value_counts()
sns.barplot(x= serie.index, y= serie.values)
plt.xlabel('Nivel')
plt.ylabel('Cantidad')
plt.title('Niveles Experencia')
```
